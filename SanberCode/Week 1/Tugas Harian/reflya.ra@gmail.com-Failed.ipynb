<h2>Quiz 2 : Built-in Function</h2>

Jawab Pertanyaan di bawah ini :
    
Apa itu built-in Function?

Sebutkan 3 built-in Function di pyhon!

Berikan contoh masing2 1 cara penggunaan built-in function yang kamu sebutkan sebelumnya

Built-in Function merupakan fungsi yang sudah disediakan, dan disupport oleh bahasa pemrograman python tanpa harus melakukan import library atau pembuatan fungsi lagi secara manual. Hal ini memudahkan pengguna python melakukan operasi-operasi di python

Contoh Build in function


1.   help 
     fungsi help merupakan build in function yang berguna untuk melihat dokumentasi singkat dari fungsi lain berupa parameter yang dimasukkan, dan keterangan singkat dari kembalian fungsi tersebut.
2.   len
     len berguna untuk melihat panjang dari sebuah container seperti list, dictionary, tuple, dan kembalian dari fungsi len adalah berupa integer.
3.   type
     type berguna untuk melihat/memastikan tipe data dari suatu objek yang dimasukkan kedalamnya, sehingga pengguna tidak salah mengidentifikasi dan melakukan operasi yang tidak seharusnya pada objek yang di identifikasi oleh type tersebut.



```python
# Contoh penggunaan built in function

a = [1, 5, 9, 12, 15, 17, 19]

# Built in function help
print("BERIKUT ADALAH CONTOH PENGGUNAAN HELP UNTUK MELIHAT KEGUNAAN DARI HELP")
print(help(help))
print()
# Built-in function len
print("BERIKUT ADALAH CONTOH PENGGUNAAN LEN DARI LIST A")
print(len(a))
print()
# Built-in function type
print("BERIKUT ADALAH CONTOH PENGGUNAAN TYPE UNTUK MELIHAT TYPE DATA DARI A")
print(type(a))
```

<h2>Quiz 3 : Method dan Function</h2>

Jawab Pertanyaan di bawah ini :
    
- Apa perbedaan method dan function?

  Beda dari method dan function ada pada objek penggunaannya, dimana pada dasarnya
  - Method adalah fungsi yang digunakan pada objek yang spesifik dan merupakan operasi yang digunakan untuk mengoperasikan objek tersebut, dan biasanya cara untuk memanggil method adalah menggunakan dot (.) setelah objek, seperti 
  (list.append()) jadi list memiliki method append untuk menggabungkan objek ke list tersebut, sementara

  - Function merupakan fungsi yang dibuat dengan tujuan untuk memberikan kembalian atau tidak memberikan kembalian dengan memastikan fungsi tersebut dipanggil sebelum objek sebagai parameter fungsi tersebut. Contohnya sorted(list, reverse=False)
  maka pada kasus ini, list adalah parameter dari fungsi sorted, dan kembalian dari fungsi ini adalah list yang sudah tersortir, tapi tidak hanya terbatas pada list tersebut saja atau objek tersebut saja, fungsi ini bisa dipergunakan pada objek lain seperti dictionary. Sementara method list.append() mungkin hanya bisa digunakan pada list saja, dengan implementasi yang dikhususkan pada list.
