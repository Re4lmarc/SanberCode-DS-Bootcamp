```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

df = pd.read_csv('https://blog.sanbercode.com/wp-content/uploads/2020/04/vgsales.csv')
df.head()
```

<h2>Quiz 1 Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
group_genre = df.groupby(df['Genre']).mean()
print(group_genre[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']])

group_genre_new = group_genre[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
group_genre_new_stack = group_genre_new.copy()
group_genre_new = group_genre_new.T
```

```python
sns.set(style='darkgrid', context='notebook', palette=sns.color_palette("bright", n_colors=12, desat=.5), color_codes=True)
```

<h2>Perbandingan Kuantitatif Barplot : Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
ax = group_genre_new.plot.bar(figsize=(18,8), width=0.8, rot=0)

plt.legend(bbox_to_anchor=(1.065, 1.165))
plt.ylabel('Mean Sales')
plt.xlabel('Region Sales')
plt.title('Mean Sales video Games By Genre')
plt.show()
```



1.   North America(NA) memiliki rata-rata penjualan paling tinggi untuk hampir semua tipe game, lalu disusul oleh Europe(EU), Japan(JP), dan Other.
2.   North America memiliki preferensi paling tinggi pada game jenis platformer, Eropa dan Other paling menyukai Shooter, dan Jepang paling menyukai Role Playing Game.
3.   Rata Rata di hampir semua negara game Adventure adalah yang memiliki penjualan cukup rendah.
4.   Hampir di semua negara, game fighting, simulation, dan sports memiliki penjualan paling konsisten dari segi tingkat penjualan.
5.   Negara Jepang kurang menyukai game fighting dan shooter
