<h2>Quiz 1 : Penggabungan Data</h2>

- Sebutkan perbedaan antara full outer join, inner join, dan left join

Jawaban :

- **Full outer join** merupakan proses join untuk semua data, dimana tidak mengharuskan tabel kiri dan kanan memiliki korelasi berupa id, code atau apapun. Jadi hasil dari join akan berupa semua data yang terkumpul. Tapi berbeda dengan cross join, setiap row tidak akan di join dengan setiap row. Hanya data yang memiliki korelasi seperti id akan digabungkan, namun untuk data yang tidak memiliki korelasi seperti id atau code yang sama akan diisi dengan data kosong atau null kecuali pada data tersebut dilakukan proses seperti smoothing.
- **inner join** merupakan gabungan antara 2 tabel yang saling memiliki korelasi saja. Jadi untuk data yang tidak memiliki korelasi, kemungkinan besar tidak akan ditunjukkan pada hasil join tersebut.
- **Left Join** merupakan proses join yang memiliki kemiripan dengan full outer join, yang membedakannya adalah datanya akan berupa full outer join - outer right. Jadi semua data yang memiliki korelasi akan di tampilkan, lalu semua data yang tidak memiliki korelasi di tabel kiri akan tampil juga, namun tidak untuk data yang tidak memiliki korelasi di tabel kanan.

<h2>Quiz 3 : Penggabungan Data</h2>

Lakukan jenis penggabungan data full outer join, inner join, left outer join dan right outer join pada kedua data berikut : 

```python
import pandas as pd

data1 = {
    'id' : [1, 2, 3, 4],
    'Feature1' : ['a1', 'a2', 'a3', 'a4'],
    'Feature2' : ['b1', 'b2', 'b3', 'b4']
}

data2 = {
    'id' : [3, 4, 5 ,6, 7],
    'Feature3' : ['c1', 'c2', 'c3', 'c4', 'c5']
}

df1= pd.DataFrame(data1)
df2= pd.DataFrame(data2)

print(df1)
print()
print(df2)
```

```python
# join data
df_outer_join = pd.merge(df1, df2, on='id',how='outer')
print(df_outer_join, "\n")

df_inner_join = pd.merge(df1, df2, on='id',how='inner')
print(df_inner_join, "\n")

df_left_join = pd.merge(df1, df2, on='id',how='left')
print(df_left_join, "\n")

df_right_join = pd.merge(df1, df2, on='id',how='right')
print(df_right_join, "\n")
```
